##Data
The dataset is from Instacart. Instacart is a grocery delivery service. Each row of the data is all of the orders from a specific department for each hour in a 24-hour day. There are 4 columns in the data they are:

- order_hour_of_day: This is the hour of the day that the order was placed
- department: This is the department of the grocery store that the product was purchased from
- num_orders_hour: This is the number of products ordered in that hour from this department
- Tot_orders_dept: This is all the products ordered from this department in a 24-hour day
